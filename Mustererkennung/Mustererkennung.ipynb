# Mustererkennung
## Einleitung
### Definition - Mustererkennung
Unter Mustererkennung verstehen wir die automatische Transformation eines Sensorsignals in eine aufgabenspezifische symbolische Beschreibung.
### Abgrenzung

***Abgrenzung zu EDV***

Aufwendige Berechnungsn werden in kurzer Zeit von einem Computer erledigt:
 - Matrixinvertiertung, Finite Elemente, Operations Research
 - Prozseesimulation in Chemie
 - World Wide Web

&rarr; Massenspeicher und hohe Rechenleistung ermöglichen Durchsatz.

***Abgrenzung zu KI***

Menschlich kinderleichte Denkprozesse und Wahrnehmungsaufgaben erfordern enorme Rechenleistung, wenn sie überhaupt machbar sind:
 - Maschineller Beweiser
 - Schachspielautomaten
 - Medizinische Diagnoseprogramme
 
&rarr; Versuch der Simulation intelligenter Verhaltensweisen. 

***Erweiterung zu Sensordatenanalyse***

Mustererkennung eher für die Simulation menschlicher Wahrnehmungsleistungen:
 - Zeichenerkennung und Dokumentenanalyse
 - Rechnersehen für autonome Fahrzeuge
 - Automatisches verstehen gesprochener Sprache
Die Sensordatenanalyse dient eher den Bereichen, in denen keine menschlichen Wahrnehmungsleistungen simuliert werden:
 - Analyse von EKG und EEG Signalen
 - Ultraschallbasierte Werkstoffprügung
 - Auswertung von Mehrkanalluftbildern
